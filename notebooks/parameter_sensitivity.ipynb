# Parameter sensitivity exploration

Interactive exploration of the parameter-sensitivity sweep behind Figure 3.
Run `experiments/parameter_sweep.py` first to populate `results/sweep.npz`.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

data = np.load('../results/sweep.npz')
list(data.keys())
```

## tau-sweep visualisation

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(data['tau_grid'], data['tau_n_voxels'], 'o-', color='C0')
ax.set_xlabel(r'$\tau$')
ax.set_ylabel('Voxel count', color='C0')
ax2 = ax.twinx()
ax2.plot(data['tau_grid'], data['tau_rmse'], 's-', color='C1')
ax2.set_ylabel('RMSE (m)', color='C1')
plt.show()
```

## (alpha, beta) heatmap exploration

Try identifying the operating-plateau region where compression and F1 are jointly favourable.

```python
fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(data['ab_compression'].T, origin='lower',
               extent=(data['alpha_grid'].min(), data['alpha_grid'].max(),
                       data['beta_grid'].min(), data['beta_grid'].max()),
               cmap='YlOrRd', aspect='auto')
ax.scatter([0.6], [0.3], marker='*', s=200, c='black', label='Working point')
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$\beta$')
plt.colorbar(im, ax=ax, label='Compression (%)')
ax.legend()
plt.show()
```
